```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths for the two CSV files
memory_controller_file = 'memory_controller/result.csv'
normal_controller_file = 'normal_controller/result.csv'

# Read the data from the CSV files into pandas DataFrames
df_memory = pd.read_csv(memory_controller_file)
df_normal = pd.read_csv(normal_controller_file)

# Create a list of data to plot
data_to_plot = [df_memory['average_gain'], df_normal['average_gain']]

# Create a boxplot with different colors for the boxes
boxplot = plt.boxplot(data_to_plot, labels=['Memory Controller', 'Normal Controller'], patch_artist=True)

# Define box colors
box_colors = ['lightblue', 'lightgreen']

# Set the box colors
for box, color in zip(boxplot['boxes'], box_colors):
    box.set_facecolor(color)

# Add labels and a title
plt.xlabel('Controller Type')
plt.ylabel('Average Gain')
plt.title('Average Gains: Enemy 8')

# Show the plot
plt.show()
```
